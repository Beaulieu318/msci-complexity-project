# Patterning Analysis

There is a range of analysis that can be done on the data. Below we show some of the basic investigations such as path length distributions and duration distributions.

## Path Length

In network terms, a path is defined as a sequence of vertices (in this case locations in the shopping mall). Each vertex is connected by an edge. The path length is defined as the sum of the lengths of each edge in the path. 

Below is a histogram of the path lengths for each mac address (left) and a processed visualisation of essentially the same data (right): log binned histogram scatter.

```python
pl,lb = lf.path_length_analysis('p')
his = lf._histogram(pl)
x = lb[0]
y = lb[1]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))
axes[0].hist(pl,bins=50)
axes[0].set_title('Histogram of Path Lengths')
axes[0].set_xlabel('Path Length')
axes[0].set_ylabel('Count')
axes[0].set_xscale('log')
axes[0].set_yscale('log')

axes[1].scatter(his[0],his[1])
axes[1].plot(x[20:],y[20:],color = 'b')
axes[1].set_title('Log Binned Path Length Probability')
axes[1].set_xlabel('Path Length')
axes[1].set_ylabel('Probability');
axes[1].set_xscale('log')
axes[1].set_yscale('log')
```
