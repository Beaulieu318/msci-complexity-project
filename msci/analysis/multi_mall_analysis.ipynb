## Import data

```python
malls = ['Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure']

mauritius_df = utils.import_signals(malls[0], version=1)
phoenix_df = utils.import_signals(malls[1], v1=True)
home_and_leisure_df = utils.import_signals(malls[2], v1=True)
```

## Devices visiting many stores

```python
from matplotlib_venn import venn3

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(20, 8))

set1s = set(mauritius_df.mac_address.tolist())
set2s = set(phoenix_df.mac_address.tolist())
set3s = set(home_and_leisure_df.mac_address.tolist())


venn3(
    [set1s, set2s, set3s], 
    ('Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure'), 
    ax=ax1
)

ax1.set_title('Venn diagram showing that the devices visit many mall');
```

A significant amount of devices visit many malls.
