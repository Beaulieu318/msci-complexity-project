## Import raw data

```python
malls = ['Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure']

mauritius_df = utils.import_signals(malls[0], version=1)
phoenix_df = utils.import_signals(malls[1], version=1)
home_and_leisure_df = utils.import_signals(malls[2], version=1)
```

## Devices visiting many stores

```python
from matplotlib_venn import venn3

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(20, 8))

set1s = set(mauritius_df.mac_address.tolist())
set2s = set(phoenix_df.mac_address.tolist())
set3s = set(home_and_leisure_df.mac_address.tolist())


venn3(
    [set1s, set2s, set3s], 
    ('Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure'), 
    ax=ax1
)

ax1.set_title(
    'Venn diagram showing the devices which have visited shopping malls'
);
```

A significant amount of devices visit many malls.

## Import cleaned data

```python
malls = ['Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure']

mauritius_df = utils.import_signals(malls[0], version=3, signal_type=1)
phoenix_df = utils.import_signals(malls[1], version=3, signal_type=1)
home_and_leisure_df = utils.import_signals(malls[2], version=3, signal_type=1)
```

## Devices visiting many stores

```python
from matplotlib_venn import venn3

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(20, 8))

set1s = set(mauritius_df.mac_address.tolist())
set2s = set(phoenix_df.mac_address.tolist())
set3s = set(home_and_leisure_df.mac_address.tolist())


venn3(
    [set1s, set2s, set3s], 
    ('Mall of Mauritius', 'Phoenix Mall', 'Home & Leisure'), 
    ax=ax1
)

ax1.set_title(
    'Venn diagram showing the devices which have visited shopping malls \
        \n (all these devices have been classified as shoppers)'
);
```

The mall (`Mall of Mauritius` and `Home & Leisure`) records an average of 700,000 visitors per month ([article](http://www.atterbury.co.za/bagatelle-mall-mauritius-celebrating-5-years/)). The mall is open 7 days a week. Therefore, this evalulates to approximately 20,000 people a day. With `Mall of Mauritius` and `Home & Leisure`, which are next to each other, this number of devices (people) that have been tracked is 15,000. Therefore, the approximate percentage of people being tracked is around 75%.
