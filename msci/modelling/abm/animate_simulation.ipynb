## Setup the environment

The environement is an object which contains an area. The area is a list of lists.

```python
environment = Environment(area=[
    ['-' for _ in range(50)] for _ in range(11)
])
```

## Create shops

```python
shops = [
    Shop('1', [(10, 4), (10, 3), (0, 3), (0,10), (10, 10), (10, 8)]),
    Shop('2', [(40, 4), (40, 3), (49, 3), (49,10), (40, 10), (40, 8)]),
]
```

```python
for shop in shops:
    shop.add_walls(environment)
```

```python
environment.output()
```

## Run the simulation

The simulation requires the maximum number of iterations (time) and the maximum number of shoppers (agents).

```python
max_iterations = 1000
max_shoppers = 4

# max_iterations = 1000
# max_shoppers = 10

shoppers_history = simulate(max_iterations, max_shoppers, environment)
```

## Animation

```python
anim = ShoppersAnimation(shoppers_history, environment, interval=100)
HTML(anim.run().to_html5_video())
```
